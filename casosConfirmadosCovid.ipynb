## Casos Confirmados de COVID-19

```python
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
urlretrieve(url, 'datasets/global_cases_covid19.csv')
```

```python
df_covid = pd.read_csv('datasets/global_cases_covid19.csv')
df_covid
```

#### Excluir latitude e longitude

```python
df_covid.drop(['Lat', 'Long'], axis=1, inplace=True)
df_covid.head()
```

```python
df_country = df_covid.groupby('Country/Region').sum()
df_country.head()
```

```python
s_brazil = df_country.loc['Brazil']
s_brazil
```

```python
date = s_brazil.index
cases = s_brazil.values
```

```python
date
```

```python
cases
```

```python
plt.figure(figsize=(14, 8))
plt.bar(date, cases)
```

```python
s_brazil = s_brazil[s_brazil > 0]
s_brazil
```

```python
tam = len(s_brazil)

plt.figure(figsize=(14, 8))
plt.xticks(rotation=60)
plt.bar(s_brazil.index[tam-30:tam], s_brazil.values[tam-30:tam])
plt.title('Último mês | Casos confirmados de COVID-19 no Brasil')
```

```python
plt.figure(figsize=(14, 8))
plt.xticks(rotation=60)
plt.bar(s_brazil.index[0:30], s_brazil.values[0:30])
plt.title('Primeiro mês | Casos confirmados de COVID-19 no Brasil')
```
